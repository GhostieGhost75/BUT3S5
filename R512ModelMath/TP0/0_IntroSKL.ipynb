## Libraries

````python
import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error

import seaborn as sns
import matplotlib.pyplot as plt
````

## Load Data

````python
df = pd.read_csv('./data/diabetes.csv')
````

````python
df
````

## Train-Test Split

````python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Outcome"]), df.Outcome, test_size=0.2) #sépare la data en deux set dont j'ai défini la taille en fonction du set de test
print(X_train)
print(X_train)
````

## Data Analysis

````python
X_train.duplicated().sum() #retourne le nombre de valeurs dupliquées
````

````python
X_train.isna().values.sum() #retourne le nombre de valeurs manquantes
````

````python
(X_train == 0).sum() #retourne le nombre de valeurs égales à 0 par colonne
````

````python
colnames = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for x in colnames :
    meancol = X_train.loc[X_train[x] != 0, x].mean()
    print(meancol)
    if x!= 'BMI' :
        meancol = round(meancol)
    X_train.loc[X_train[x] == 0, x] = round(meancol, 1)
print(X_train)
````

````python
plt.figure(figsize=(10, 4))
sns.countplot(y=y_train, order=y_train.value_counts().index)
plt.xticks(rotation = 15)
plt.show()
````

````python
plt.figure(figsize=(10, 4))
plt.scatter(X_train.index, X_train.Glucose, c=y_train)
plt.xticks(rotation = 15)
ax = plt.gca()
ax.set_facecolor('grey')
plt.show()
````
